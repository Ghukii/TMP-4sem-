Использовать механизм сводных таблиц (метод `pivot_table()`).

```python
t = tr[['store_id', 'store_sales', 'product_id']].merge(pr[['product_id', 'brand_name']], on='product_id', how='inner')
pivot_table = pd.pivot_table(t, values='store_sales', index='store_id', columns='brand_name', aggfunc='mean')
pivot_table
# Создаем таблицу t путем объединения таблиц tr и pr по столбцу product_id.
# Создаем сводную таблицу pivot_table с помощью метода pivot_table библиотеки pandas.
# В качестве значений используем столбец store_sales, в качестве индекса - store_id, в качестве столбцов - brand_name.
```

```python
t = tr[['store_id', 'store_sales', 'product_id', 'month_of_year']].merge(pr[['product_id', 'brand_name']], on='product_id', how='inner')
cross_table = pd.crosstab(index=[t['brand_name']], columns=t['month_of_year'], values=t['store_sales'], aggfunc='sum')
cross_table
# Создаем таблицу t путем объединения таблиц tr и pr по столбцу product_id.
# Создаем сводную таблицу cross_table с помощью метода crosstab библиотеки pandas.
# В качестве индекса используем столбец brand_name, в качестве столбцов - month_of_year.
# В качестве значений используем столбец store_sales, в качестве функции агрегации - сумма.
```
