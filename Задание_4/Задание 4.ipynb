Определить среднюю стоимость покупок в различных магазинах ('store_id') для продуктов различных брендов ('brand_name').

Использовать механизм сводных таблиц (метод `pivot_table()`).

Объяснить, что означают `NaN` в полученной результирующей таблице.

```python
tmp = transactions[['store_id', 'store_sales', 'product_id']].merge(products[['product_id', 'brand_name']], on='product_id', how='inner')
# Объединяем DataFrame `transactions`, содержащий 'store_id', 'store_sales', 'product_id',
# с DataFrame `products`, содержащим 'product_id' и 'brand_name'.
# Соединение происходит по 'product_id', тип соединения - внутреннее (inner), чтобы сохранить только те строки,
# в которых существует соответствие в обоих DataFrame.

pivot_table = pd.pivot_table(tmp, values='store_sales', index='store_id', columns='brand_name', aggfunc='mean')
# `pivot_table` используется для агрегации средней стоимости продаж (`store_sales`) по каждому магазину (`store_id`)
# и по каждому бренду (`brand_name`), что помогает анализировать средние продажи брендов в разных магазинах.

pivot_table # Вывод сводной таблицы
```

ВЫВОД: Если в исходном DataFrame нет записей о покупках для конкретного бренда в конкретном магазине, в соответствующей ячейке таблицы будет стоять значение NaN.

Определить количество продаж в каждом месяце ('month_of_year') для каждого бренда ('brand_name').

Использовать кросс-таблицы.

Пояснить, что означают числа в ячейках полученной таблицы.

```python
tmp = transactions[['store_id', 'store_sales', 'product_id', 'month_of_year']].merge(products[['product_id', 'brand_name']], on='product_id', how='inner') # Создаем таблицу tmp путем объединения таблиц transactions и products по столбцу product_id.

cross_table = pd.crosstab(index=[tmp['brand_name']], columns=tmp['month_of_year'], values=tmp['store_sales'], aggfunc='sum')
# Создаем сводную таблицу cross_table с помощью метода crosstab библиотеки pandas.
# В качестве индекса используем столбец brand_name, в качестве столбцов - month_of_year.
# В качестве значений используем столбец store_sales, в качестве функции агрегации - сумма.

cross_table # Вывод DataFrame
```

ОБЪЯСНЕНИЕ: Числа в ячейках полученной кросс-таблицы представляют собой количество продаж конкретного бренда в каждом месяце. Например, если в ячейке на пересечении строки с брендом 'Amigo' и столбца с месяцем '6' стоит число 159.98, это означает, что в январе было продано 159.98 единиц продукции бренда 'Amigo'.

Сформировать таблицу df1, в которой будут записи о транзакциях тех групп, в которых средняя стоимость покупок каждого покупателя составляет не менее 13.

Использовать метод `filter()`.

```python
tmp = transactions[['customer_id', 'store_sales']].groupby('customer_id').mean().reset_index()
# Выборка определенных столбцов из DataFrame `transactions` и группировка по 'customer_id'.
# Затем для каждой группы рассчитывается среднее значение 'store_sales'.
# После выполнения группировки и агрегации выполняется сброс индекса для возврата DataFrame.

df1 = tmp[tmp['store_sales'] >= 13]
# Фильтрация сформированного DataFrame, чтобы оставить только те строки,
# где средние 'store_sales' больше или равны 13.

df1 # Вывод отфильтрованного DataFrame
```

Добавить в таблицу df столбец 'store_sales_group'
с категорией суммы покупки:
- `small` - сумма покупки < 10;
- `average` - сумма покупки >= 10 и < 13;
- `large` - сумма покупки >= 13.

Использовать функцию `applay()`.

```python
def check(x): # Определение функции check для классификации значений в столбце store_sales
    if x < 10:
        return 'small'
    elif x >= 10 and x < 13:
        return 'average'
    return 'large'

df['store_sales_group'] = df['store_sales'].apply(check) # Применение функции check к столбцу store_sales с использованием метода apply()

df # Вывод обновленного DataFrame
```

Для выполнения следующих заданий используется набор данных о динамике велосипедного движения между центром города и некоторым его районом:
- `Time` - время измерения;
- `direction_centre`	- количество велосипедистов в направлении центра;
- `direction_region` - количество велосипедистов в направлении района.

Считать файл 'time_series.csv' и записать его в таблицу 'dft'.

```python
dft = pd.read_csv('time_series.csv') # Читаем файл time_series.csv с помощью метода read_csv
dft.head() # Выводим первые 5 строк таблицы с помощью метода head.
```

На основе информаци о дате и времени создать в таблице dft столбец 'Date_Time'.

Элемент нового столбца должен иметь следующий вид:
14/07/2011	17:44

```python
dft['Date_Time'] = dft['Date'] + ' ' + dft['Time'] # Создаем новый столбец Date_Time в таблице dft
dft.head() # Выводим первые 5 строк таблицы с помощью метода head.
```

Удалить столбцы 'Date' и 'Time'.

```python
dft = dft.drop(columns=['Date', 'Time']) # Удаляем столбцы Date и Time из таблицы dft с помощью метода drop
dft.head() # Выводим первые 5 строк таблицы с помощью метода head
```

Преобразовать тип переменной 'Date_Time' в формат `datetime`.


```python
dft['Date_Time'] = pd.to_datetime(dft['Date_Time'], format='%d/%m/%Y %H:%M') # Преобразуем столбец Date_Time в тип datetime с помощью метода to_datetime
dft.head() # Выводим первые 5 строк таблицы с помощью метода head.
```

Установить столбец 'Date_Time' в качестве индекса.

```python
dft = dft.set_index('Date_Time') # Устанавливаем столбец Date_Time в качестве индекса таблицы df с помощью метода set_index.
dft.head() # Выводим первые 5 строк таблицы с помощью метода head.
```

Добавить признак 'both' с суммарным потоком велосипедистов.

```python
dft = dft.assign(both=dft['direction_centre'] + dft['direction_region']) # Создаем новый столбец both в таблице dft, соединяя значения столбцов direction_centre и direction_region
dft.head() # Выводим первые 5 строк таблицы с помощью метода head
```

Добавить столбец 'weekday' с названиями дней недели на русском языке.

```python
dft['weekday'] = dft.index.day_name() # Создаем новый столбец weekday в таблице dft, содержащий название дня недели для каждой строки.

english_to_russian = { # Создаем словарь english_to_russian, содержащий перевод названий дней недели с английского на русский язык
    'Monday': 'Понедельник',
    'Tuesday': 'Вторник',
    'Wednesday': 'Среда',
    'Thursday': 'Четверг',
    'Friday': 'Пятница',
    'Saturday': 'Суббота',
    'Sunday': 'Воскресенье'
}

dft['weekday'] = dft['weekday'].map(english_to_russian) # Применяем метод map к столбцу weekday, чтобы заменить названия дней недели на русский язык.
dft.head() # Выводим первые 5 строк таблицы с помощью метода head
```

Выбрать временные интервалы, для которых суммарный поток менее 3 велосипедистов.

Выбрать временные интервалы, для которых в одном из направлений проехало менее 5 велосипедистов.

```python
dft[(dft['direction_centre'] < 5) | (dft['direction_region'] < 5)] # Выбираем строки из таблицы dft, в которых значение столбца direction_centre меньше 5
```

Определить среднее количество велосипедистов за 15 минут.

```python
dft['total'] = dft['direction_centre'] + dft['direction_region'] # Создаем новый столбец total в таблице dft, содержащий сумму значений столбцов direction_centre и direction_region
dft['total'].mean() # Вычисляем среднее значение столбца total с помощью метода mean.
```

Определить среднее количество велосипедистов за 1 час.

```python
dft['total'].resample('h').sum().mean()
# Группируем данные по часам с помощью метода resample.
# Суммируем значения столбца total в каждой группе с помощью метода sum.
# Вычисляем среднее значение суммы столбца total в каждой группе с помощью метода mean.
```

Найти даты с 5-ю самыми большими значениями количества велосипедистов, проехавших в направлении к центру.
