````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

### **Задание 1**

Построить в одном окне график синусоиды и график прямой линии.
Графики должны иметь разные цвета, стили линий и тип маркеров.
Подписать название окна, названия осей.
Вывести легенду.

````python
# Создаем массив x_sin равномерно распределенных точек от 0 до 10, всего 100 точек
x_sin = np.linspace(0, 10, 100)
# Вычисляем значения синуса для каждой точки массива x_sin
y_sin = np.sin(x_sin)

# Создаем аналогичный массив x_pr для визуализации линейной функции
x_pr = np.linspace(0, 10, 100)
# Расчитываем y_pr как линейную функцию от x_pr: y = 0.3x - 1
y_pr = 0.3 * x_pr - 1

# Строим линейный график для синусоиды с помощью пунктирной линии красного цвета
plt.plot(x_sin, y_sin, linestyle='--', color='red', zorder=0, label='Синусоида')

# Размещаем точки на графике синусоиды через каждые 10 элементов с кружками красного цвета
plt.scatter(x_sin[::10], y_sin[::10], marker='o', color='red', edgecolor='black', linewidth=2, s=75, zorder=1)

# Помечаем последнюю точку на графике синусоиды
plt.scatter(x_sin[-1], y_sin[-1], marker='o', color='red', edgecolor='black', linewidth=2, s=75, zorder=1)

# Строим линейный график для прямой с помощью тире-точка линии зеленого цвета
plt.plot(x_pr, y_pr, linestyle='-.', color='green', zorder=0, label='Прямая')

# Размещаем звездочки зеленого цвета на прямой через каждые 25 элементов
plt.scatter(x_pr[::25], y_pr[::25], marker='*', color='green', edgecolor='black', linewidth=1, s=150, zorder=1)

# Помечаем последнюю точку на графике прямой
plt.scatter(x_pr[-1], y_pr[-1], marker='*', color='green', edgecolor='black', linewidth=1, s=150, zorder=1)

# Создаем объекты для легенды, которые соответствуют стилям на графике
sin_legend = plt.Line2D([0], [0], linestyle='--', color='red', marker='o',
                        markersize=7.5, markeredgecolor='black', markeredgewidth=2, label='Синусоида')
pr_legend = plt.Line2D([1], [0], linestyle='-.', color='green', marker='*', markersize=10, markeredgecolor='black', 
                       markeredgewidth=1.2, label='Прямая')

# Добавляем легенду на график
plt.legend(handles=[sin_legend, pr_legend])

# Устанавливаем заголовок графика и подписи к осям
plt.title('График прямой и синусоиды')
plt.xlabel('X')
plt.ylabel('Y')

# Отображаем готовый график
plt.show()
````

### **Задание 2**

Построить 5 субграфиков различных линий и разместить их в области рисования следующего вида:

```
|-------------------------|
|        График 1         |
|-------------------------|
|  График 2  |  График 3  |
|-------------------------|
|  График 4  |  График 5  |
|-------------------------|
```

````python
# Генерация данных для графиков
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.sinh(x)
y5 = np.cosh(x)

# Создание области рисования (фигуры) размером 10x8 
fig = plt.figure(figsize=(10, 8))

# График 1: Синус (верхний большой график)
ax1 = plt.subplot2grid((3, 2), (0, 0), colspan=2)  # Размещение на сетке 3x2, занимает два столбца
ax1.plot(x, y1, 'red')  # Построение графика синуса с красной линией
ax1.set_title('Синус')  # Установка заголовка графика

# График 2: Косинус (второй ряд, первый субграфик)
ax2 = plt.subplot2grid((3, 2), (1, 0))  # Размещение на сетке 3x2 во втором ряду, первом столбце
ax2.plot(x, y2, 'green')  # Построение графика косинуса с зеленой линией
ax2.set_title('Косинус')  # Установка заголовка графика

# График 3: Тангенс (второй ряд, второй субграфик)
ax3 = plt.subplot2grid((3, 2), (1, 1))  # Размещение на сетке 3x2 во втором ряду, втором столбце
ax3.plot(x, y3, 'blue')  # Построение графика тангенса с синей линией
ax3.set_title('Тангенс')  # Установка заголовка графика

# График 4: Гиперболический синус (третий ряд, первый субграфик)
ax4 = plt.subplot2grid((3, 2), (2, 0))  # Размещение на сетке 3x2 в третьем ряду, первом столбце
ax4.plot(x, y4, 'purple')  # Построение графика гиперболического синуса с пурпурной линией
ax4.set_title('Гиперболический синус')  # Установка заголовка графика

# График 5: Гиперболический косинус (третий ряд, второй субграфик)
ax5 = plt.subplot2grid((3, 2), (2, 1))  # Размещение на сетке 3x2 в третьем ряду, втором столбце
ax5.plot(x, y5, 'black')  # Построение графика гиперболического косинуса с черной линией
ax5.set_title('Гиперболический косинус')  # Установка заголовка графика

# Настройка отступов между субграфиками для улучшения читаемости
plt.tight_layout()

# Отображение графиков
plt.show()
````

### **Задание 3**

#### 3.1. Создать таблицу, содержащую 3 столбца:
– «P1» – количественный признак со случайными вещественными значениями в диапазоне от 0 до 1;
– «P2» – количественный признак со случайными целыми значениями в диапазоне от 1 до 100;
– «P3» – качественный признак с тремя значениями «a», «b» и «с».

````python
n = 30  # количество строк

# Создаем DataFrame
df = pd.DataFrame({
    'P1': np.random.rand(n),  # случайные вещественные числа от 0 до 1
    'P2': np.random.randint(1, 101, size=n),  # случайные целые числа от 1 до 100
    'P3': np.random.choice(['a', 'b', 'c'], size=n)  # случайный выбор между 'a', 'b', 'c'
})
````

#### 3.2. Построить график зависимости признака «P2» от «P1». Различными цветами отобразить принадлежность объектов к категориям «a», «b» и «с».
Подписать название графика, названия осей.
Вывести легенду. 

````python
# Группируем данные по категориям из столбца 'P3'
groups = df.groupby('P3')

# Создаем график
plt.figure(figsize=(10, 6))

# Проходимся по каждой категории и строим соответствующие точки на графике
for name, group in groups:
    plt.scatter(group['P1'], group['P2'], label=name)  # точки с соответствующим цветом для каждой категории

# Добавляем подписи осей и название графика
plt.title('Зависимость признака P2 от P1 с различными категориями')
plt.xlabel('P1')
plt.ylabel('P2')

# Выводим легенду
plt.legend()

# Отображаем график
plt.show()
````

#### 3.3. Построить гистограммы признака «P1» для различных значений признака «P3».
Графики должны иметь разные цвета.
Подписать название графика, названия осей.
Вывести легенду. 

````python
# Группируем данные по категориям из столбца 'P3'
groups = df.groupby('P3')

# Создаем гистограммы для каждой категории
plt.figure(figsize=(10, 6))
for name, group in groups:
    plt.hist(group['P1'], bins=15, alpha=0.5, label=name)  # создаем гистограмму для каждой категории с разным цветом

# Добавляем подписи осей и название графика
plt.title('Гистограммы признака P1 для различных значений P3')
plt.xlabel('P1')
plt.ylabel('Частота')

# Выводим легенду
plt.legend()

# Отображаем график
plt.show()
````

#### 3.4. Построить диаграмму, отображающую количество значений признака «P2» и «P1» для различных значений признака «P3».
Графики должны иметь разные цвета.
Подписать название графика, названия осей.
Вывести легенду.

````python
# Агрегируем данные по 'P3' для получения суммы 'P1' и среднего 'P2'
aggregated_data = df.groupby('P3').agg({
    'P1': 'count',
    'P2': 'count'
}).reset_index()

# Создаем столбчатую диаграмму
fig, ax = plt.subplots(figsize=(10, 6))
width = 0.35  # ширина столбцов

# Столбцы для 'P1'
rects1 = ax.bar(aggregated_data['P3'], aggregated_data['P1'], width, label='Кол-во P1', color='blue')

# Столбцы для 'P2'
rects2 = ax.bar(aggregated_data['P3'], aggregated_data['P2'], width, label='Кол-во P2', color='green', bottom=aggregated_data['P1'])

# Добавляем подписи осей и название графика
ax.set_title('Количество значений для P1 и P2 по категориям P3')
ax.set_xlabel('Категория P3')
ax.set_ylabel('Значения')
ax.legend()

# Отображаем график
plt.show()
````

#### 3.5 Построить диаграмму «ящик с усами» для признака «P2» при различных значениях признака «P3».
Графики должны иметь разные цвета.
Подписать название графика, названия осей.
Вывести легенду.

````python
# Создаем список цветов для каждой категории
colors = {'a': 'red', 'b': 'green', 'c': 'blue'}

# Создаем список данных для каждой категории
data = [df[df['P3'] == p]['P2'] for p in df['P3'].unique()]

# Создаем диаграмму "ящик с усами"
plt.figure(figsize=(10, 6))

# Построение ящиков с усами с разными цветами
plt.boxplot(data, patch_artist=True, labels=df['P3'].unique())

# Добавляем подписи осей и название графика
plt.title('Диаграмма "ящик с усами" для признака P2 по категориям P3')
plt.xlabel('Категория P3')
plt.ylabel('Значения P2')

# Создаем легенду
legend_labels = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in df['P3'].unique()]
plt.legend(legend_labels, df['P3'].unique())

# Отображаем график
plt.show()
````
