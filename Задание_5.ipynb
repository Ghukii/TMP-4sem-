### **Задание 1**

Построить в одном окне график синусоиды и график прямой линии.
Графики должны иметь разные цвета, стили линий и тип маркеров.
Подписать название окна, названия осей.
Вывести легенду.

````python
x_sin = np.linspace(0, 10, 100)  # Генерация 100 точек от 0 до π
y_sin = np.sin(x_sin) # Создание массива значений y - синусоиды

x_pr = np.linspace(0, 10, 100)
y_pr = 0.3 * x_pr - 1

# Построение графика
plt.plot(x_sin, y_sin, linestyle='--', color='red', zorder=0, label='Синусоида')
plt.scatter(x_sin[::10], y_sin[::10], marker='o', color='red', edgecolor='black', linewidth=2, s=75, zorder=1)
plt.scatter(x_sin[-1], y_sin[-1], marker='o', color='red', edgecolor='black', linewidth=2, s=75, zorder=1)
plt.plot(x_pr, y_pr, linestyle='-.', color='green', zorder=0, label='Прямая')
plt.scatter(x_pr[::25], y_pr[::25], marker='*', color='green', edgecolor='black', linewidth=1, s=150, zorder=1)
plt.scatter(x_pr[-1], y_pr[-1], marker='*', color='green', edgecolor='black', linewidth=1, s=150, zorder=1)

# Создание "псевдо-элементов" для легенды с корректными параметрами
sin_legend = plt.Line2D([0], [0], linestyle='--', color='red', marker='o',
                        markersize=7.5, markeredgecolor='black', markeredgewidth=2, label='Синусоида')
pr_legend = plt.Line2D([1], [0], linestyle='-.', color='green', marker='*', markersize=10, markeredgecolor='black', 
                       markeredgewidth=1.2, label='Прямая')

# Добавление легенды с псевдо-элементами
plt.legend(handles=[sin_legend, pr_legend])

# Добавление заголовка и подписей осей
plt.title('График прямой и синусоиды')
plt.xlabel('X')
plt.ylabel('Y')

# Отображение графика
plt.show()
````

### **Задание 2**

Построить 5 субграфиков различных линий и разместить их в области рисования следующего вида:

```
|-------------------------|
|        График 1         |
|-------------------------|
|  График 2  |  График 3  |
|-------------------------|
|  График 4  |  График 5  |
|-------------------------|
```

````python
# Генерация данных для графиков
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.sinh(x)
y5 = np.cosh(x)

# Создание области рисования
fig = plt.figure(figsize=(10, 8))

# График 1 (верхний большой график)
ax1 = plt.subplot2grid((3, 2), (0, 0), colspan=2)
ax1.plot(x, y1, 'red')
ax1.set_title('Синус')

# График 2 (второй ряд, первый субграфик)
ax2 = plt.subplot2grid((3, 2), (1, 0))
ax2.plot(x, y2, 'green')
ax2.set_title('Косинус')

# График 3 (второй ряд, второй субграфик)
ax3 = plt.subplot2grid((3, 2), (1, 1))
ax3.plot(x, y3, 'blue')
ax3.set_title('Тангенс')

# График 4 (третий ряд, первый субграфик)
ax4 = plt.subplot2grid((3, 2), (2, 0))
ax4.plot(x, y4, 'purple')
ax4.set_title('Гиперболический синус')

# График 5 (третий ряд, второй субграфик)
ax5 = plt.subplot2grid((3, 2), (2, 1))
ax5.plot(x, y5, 'black')
ax5.set_title('Гиперболический тангенс')

# Настройка отступов между субграфиками
plt.tight_layout()

# Отображение графиков
plt.show()
````

#### 3.5 Построить диаграмму «ящик с усами» для признака «P2» при различных значениях признака «P3».
Графики должны иметь разные цвета.
Подписать название графика, названия осей.
Вывести легенду.

````python
# Создание данных
np.random.seed(13)
data = {
    'P1': np.random.rand(100),
    'P2': np.random.randint(1, 101, size=100),
    'P3': np.random.choice(['a', 'b', 'c'], size=100)
}

# Создание DataFrame
df = pd.DataFrame(data)
df
````

````python
# Построение графика
plt.figure(figsize=(8, 6))
for category in df['P3'].unique():
    plt.scatter(df[df['P3'] == category]['P1'], df[df['P3'] == category]['P2'], label=category)

# Настройка графика
plt.title('Зависимость P2 от P1')
plt.xlabel('P1')
plt.ylabel('P2')
plt.xticks(np.arange(0, 1.2, 0.1))
plt.yticks(np.arange(0, 111, 10))
plt.legend(loc='upper right')
plt.grid(True)

# Отображение графика
plt.show()
````

````python
# Построение гистограмм
plt.figure(figsize=(8, 6))
for category in df['P3'].unique():
    plt.hist(df[df['P3'] == category]['P1'], bins=10, alpha=0.5, label=category)

# Настройка графика
plt.title('Гистограммы P1 для различных значений P3')
plt.xlabel('P1')
plt.ylabel('Частота')
plt.legend(title='P3')
plt.grid(True)

# Отображение графика
plt.show()
````

````python
# Построение диаграммы
plt.figure(figsize=(8, 6))
for category in df['P3'].unique():
    plt.bar(category, df[df['P3'] == category]['P1'].count(), label=f'{category} - P1', alpha=0.5)
    plt.bar(category, df[df['P3'] == category]['P2'].count(), label=f'{category} - P2', alpha=0.5)

# Настройка графика
plt.title('Количество значений P1 и P2 для различных значений P3')
plt.xlabel('P3')
plt.ylabel('Количество')
plt.legend()
plt.grid(True)

# Отображение графика
plt.show()
````

````python
# Построение ящика с усами
plt.figure(figsize=(8, 6))
plt.boxplot([df[df['P3'] == 'a']['P2'], df[df['P3'] == 'b']['P2'], df[df['P3'] == 'c']['P2']],
            labels=['a', 'b', 'c'])

# Настройка графика
plt.title('Ящик с усами для P2 при различных значениях P3')
plt.xlabel('P3')
plt.ylabel('P2')
plt.grid(True)

# Отображение графика
plt.show()
````
