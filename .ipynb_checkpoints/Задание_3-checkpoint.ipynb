```python
import pandas as pd # Подключаем библиотеку Pandas
import numpy as np # Подключаем библиотеку Numpy
```

#### 1. Создать объект Series на основе массива numpy. 

```python
arr = np.random.randint(-100, 100, 10) # Создаем массив рандомных чисел
s = pd.Series(arr) # Создаем Series на основе массива
s # Выводим Series
```

#### 2. Создать объект DataFrame на основе списка словарей. Количество строк – 12, количество столбцов – 3.

```python
ls = list() # Создаем список

for i in range(12):
    ls.append({'column_1': np.random.randint(10), 'column_2': chr(97 + i) * 5, 'column_3': i ** 2}) # Заполняем список словарями

df = pd.DataFrame(ls) # Создаем DataFrame
df # Выводим DataFrame
```

#### 3. Вывести информацию о количестве столбцов в таблице.

```python
print('Количество столбцов:', df.shape[1])
```

#### 4. Выбрать строки с 1 по 3 столбцов 1 и 2. Реализовать двумя способами: а) индикатор loc, б) индикатор iloc.

##### а) индикатор loc

```python
df[["column_1", "column_2"]].loc[1:3]
```

##### б) индикатор iloc

```python
df[["column_1", "column_2"]].iloc[1:4]
```

#### 5. Добавить в таблицу столбец с вещественными числами.

```python
new = [np.random.rand() * 10 for _ in range(12)] # Создаем список из 12 вещественных чисел
df.insert(0, "floats", new) # Добавляем список в DataFrame как столбец
df
```

#### 6. Переименовать 1-й столбец.

```python
df = df.rename(columns={"floats": "float_column"})
df
```

#### 7. Удалить 1-ю строку таблицы. 

```python
df = df.loc[1:]
df
```

#### 8. Заменить любые два значения в 4-м столбце на 777,7 и 888,8

```python
df = df.replace(4, 777)
df = df.replace(64, 888)
df
```

#### 9. Вывести строки таблицы, в которых значения столбца 4 меньше или равны заданному

```python
val = int(input())

df[df["column_3"] <= val]
```

#### 10. Заменить значения в столбце 4 произведением значений столбцов 1 и 2

```python
ls = [int(df["float_column"].iloc[i] * df["column_1"].iloc[i]) for i in range(df.shape[0])] # Создаем список из произведений значений столбцов 1 и 2
df = df.replace(df["column_3"].tolist(), ls) # Заменяем значения в 4 стобце значениями из списка
df
```
